## Operasi Aritmatika

+ tanda + adalah penjumlahan
+ tanda - adalah pengurangan
+ tanda / adalah pembagian
+ tanda * perkalian

```r
print('1 + 1 sama dengan')
1 + 1

print('2 - 3 sama dengan')
2 - 3

print('8/3 sama dengan')
8/3

print('9 * 0.5 sama dengan')
9*0.5

print('6 ^ 9 sama dengan')
6^9

print('9 %% 3')
9%%3
```

## Percabangan

Percabangan adalah cara yang digunakan untuk mengambil keputusan apabila di dalam program dihadapkan pada kondisi tertentu. Contoh kita ingin mencetak 'Anda lulus ujian' jika nilai yang di peroleh >= 70

if ( __kondisi__ ){ <br>
    __statment__ <br>
}

Logika nya statement akan di eksekusi jika berniali TRUE (benar)

```r
nilai = 60
```

```r
nilai > 70
```

```r
if(nilai > 70){
    print('Anda lulus ujian')
}
```

```r
if(nilai > 70){
    print('Anda lulus ujian')
}else{
    print('anda tidak lulus ujian')
}
```

Sekarang buat program:

Jika > 70, lulus ujian <br>
jika = 70, anda tidak lulus, tapi boleh remedial <br>
jika < 70 anda tidak lulus. <br>

```r
nilai  <- 70
```

```r
if(nilai > 70){
    print('anda lulus ujian')
} else if( nilai == 70){
    print('anda tidak lulus ujian, tapi boleh remedian')
} else{
    print('anda tidak lulus ujian')
}
```

## Kembali ke perulangan

```r
print(paste('9*1 =', 9*1))
print(paste('9*2 =', 9*2))
print(paste('9*3 =', 9*3))
print(paste('9*4 =', 9*4))
print(paste('9*5 =', 9*5))
print(paste('9*6 =', 9*6))
print(paste('9*7 =', 9*7))
print(paste('9*8 =', 9*8))
print(paste('9*9 =', 9*9))
print(paste('9*10 =', 9*10))
```

```r
numbers = c(1:10)

for(i in numbers){
    print(paste('9 *', i, '=', 9*i))
}
```

```r
numbers = c(1:10)
perkalian = c()
for(i in numbers){
    perkalian[i] = 9*i
}
perkalian
```

# Perulangan dengan break

```r
numbers = c(1:10)
perkalian = c()
for(i in numbers){
  perkalian[i] = 9*i
  if(i == 5){
    break
  }
}
perkalian
```

# Perulangan dengan next

```r
numbers = c(1:10)
perkalian = c()
for(i in numbers){
  if(i == 5){
    perkalian[i] = 0
    next
  }
  perkalian[i] = 9*i
}
perkalian
```

```r
numbers <- c(1:10)
for(i in numbers){
    if(i %% 2 == 0){
        print(paste(i, 'adalah bilangan genap'))
    } else{
        print(paste(i, 'adalah bilangan ganjil'))
    }
}
```

## Fungsi

Fungsi bahasa pemrograman yaitu memerintah komputer untuk mengolah data sesuai dengan alur berpikir yang kita inginkan.





### Penulisan fungsi

#### cari 1
akar <- function(x) sqrt(x)

### cara 2
akar <- function(x){
    sqrt(x)
}

+ akar adalah nama fungsi
+ function untuk mendefinisikan kalo akar itu adalah fungsi
+ x adalah inputan / parameter
+ sqrt(x) adalah proses pada fungsi

```r
### fungsi tanpa parameter

hay = function(){
    print('Hay, selamat datang di welcome')
}
```

```r
hay()
```

```r
### fungsi dengan paramater

hay_input <- function(nama){
    print(paste0('hay ', nama, ', selamat datang di welcome'))
}
```

```r
hay_input()
```

```r
hay_input('Ari')
```

```r
### fungsi dengan paramater yang di inisiasi

hay_input <- function(nama='Ari'){
    print(paste0('hay ', nama, ', selamat datang di welcome'))
}
```

```r
hay_input()
```

```r
hay_input('Kiki')
```
